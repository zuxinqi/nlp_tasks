```python
import re
import json
import pickle
```

```python
# BMES标注
def origin2tag_new(untagged_file=None):
    tagged_paragraph_list = []          # 存放标记后的数据
    tag_names = []
    for paragraph in untagged_file.readlines():
        # print(line)               # 当出现错误时，打印进行脏数据处理
        tagged_paragraph_str, i = '', 0
        while i<len(paragraph.rstrip('\n')):          # 对于每一句原始数据，逐字进行遍历
            if paragraph[i] == '{':      # 发现实体
                i+=2
                temp=""
                while paragraph[i]!='}':
                    temp+=paragraph[i]
                    i+=1
                i+=2

                tag, entity= temp.split(':')[0], temp.split(':')[1].strip()
                tag_names.append(tag)
                if len(entity)==1:  # 针对S标记
                    tagged_paragraph_str += entity+'/S_'+tag+' '
                else:               # 针对B、I、E标记
                    entity = entity.replace(' ', '_')
                    tagged_paragraph_str += entity[0]+'/B_'+tag+' '
                    for j in range(1, len(entity)-1):
                        tagged_paragraph_str += entity[j]+'/M_'+tag+' '
                    tagged_paragraph_str += entity[-1]+'/E_'+tag+' '
            else:                   # 对于非实体O标记
                tagged_paragraph_str += paragraph[i]+'/O '
                i += 1
        tagged_paragraph_list.append(tagged_paragraph_str)
    
    return tagged_paragraph_list,tag_names
```

```python
# BIO标注
def origin2tag(untagged_file=None):
    tagged_paragraph_list = []          # 存放标记后的数据
    tag_names = []
    for paragraph in untagged_file.readlines():
        # print(line)               # 当出现错误时，打印进行脏数据处理
        tagged_paragraph_str, i = '', 0
        while i<len(paragraph.rstrip('\n')):          # 对于每一句原始数据，逐字进行遍历
            if paragraph[i] == '{':      # 发现实体
                i+=2
                temp=""
                while paragraph[i]!='}':
                    temp+=paragraph[i]
                    i+=1
                i+=2

                tag, entity= temp.split(':')[0], temp.split(':')[1].strip()
                tag_names.append(tag)
                if len(entity)==1:  # 针对S标记
                    tagged_paragraph_str += entity+'/B_'+tag+' '
                else:               # 针对B、I、E标记
                    entity = entity.replace(' ', '_')
                    tagged_paragraph_str += entity[0]+'/B_'+tag+' '
                    for j in range(1, len(entity)):
                        tagged_paragraph_str += entity[j]+'/I_'+tag+' '
            else:                   # 对于非实体O标记
                tagged_paragraph_str += paragraph[i]+'/O '
                i += 1
        tagged_paragraph_list.append(tagged_paragraph_str)
    
    return tagged_paragraph_list, tag_names
```

```python
def paragraph_split(tagged_paragraph_list=None):
    sentences_list, sent_split = [], ["。/O", "！/O", "？/O", "；/O"]
    for paragraph in tagged_paragraph_list:
        temp = ''
        for sent in re.split('([ 。 ！ ？ ；]/[O])', paragraph):
            if sent not in sent_split:
                temp += sent
            else:
                if temp!='':
                    sentences_list.append(temp+sent)
                    temp = ''
                else:
                    continue
    return sentences_list
```

```python
def flatten2result(input_list=None):
    res_all_list = []
    for sentence in input_list:
        one_list = [[],[]]
        for char_tag_pair in sentence.rstrip('\n').split():
            if char_tag_pair.lstrip()[0] == '/':
                continue
            else:
                one_list[0].append(char_tag_pair.split("/")[0])
                one_list[1].append(char_tag_pair.split("/")[1])
        if len(one_list[0]) != len(one_list[1]):
            print(111111111111)
        res_all_list.append(one_list)
    return res_all_list
```

```python
with open("BosonNLP_NER_6C.txt", 'r', encoding="utf-8") as f:
    tagged_paragraph_list,tag_names = origin2tag(f)
    sentences_list = paragraph_split(tagged_paragraph_list)
    all_list = flatten2result(sentences_list)
```

```python
with open("boson_ner_all_data.pkl", 'wb') as f:
    pickle.dump(all_list, f)
```
